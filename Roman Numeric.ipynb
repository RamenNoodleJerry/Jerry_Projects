### The following code transform roman numerals to integer:

```python
values =  {
  "I": 1,
  "V": 5,
  "X": 10,
  "L": 50,
  "C": 100,
  "D": 500,
  "M": 1000,
}

def RomanToInt(str):
 total = 0
 i = 0
    
 while (i < len(str)):
  S1 = values[str[i]]
  if (i+1 < len(str)):
       S2 = values[str[i+1]]
       if (S1>=S2):
         total = total + S1
         i = i + 1
       else:    
         total = total - S1
         i = i + 1
  else:
     total = total + S1
     i = i + 1
 return total

#test code
print(RomanToInt("MCMXCIV"))
        
```
